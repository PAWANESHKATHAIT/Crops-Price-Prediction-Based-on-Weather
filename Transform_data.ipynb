```python
import pandas as pd
```

```python
df1 = pd.read_csv("crop_dataset.csv")
df1
```

```python
df2 = pd.read_csv("weather_data.csv")
df2
```

```python
merged_df = pd.merge(df1, df2, left_on='District', right_on='city', how='inner')
merged_df
```

```python
merged_df.info()
```

 ### Data Processing

```python
df = merged_df
```

```python
df = df.rename(columns={'Min_x0020_Price': 'Min_Price', 'Max_x0020_Price': 'Max_Price', 'Modal_x0020_Price': 'Frequent_price'})
```

```python
# Convert prices from quintals to kilograms by dividing by 100 and update the existing columns
df['Min_Price'] = df['Min_Price'] / 100
df['Max_Price'] = df['Max_Price'] / 100
df['Frequent_price'] = df['Frequent_price'] / 100
```

```python
df.drop(columns = ['city', 'season', 'timestamp', 'Arrival_Date'], inplace = True)
```

```python
exclude_columns = ['wind_speed']

# Convert all decimal values to integers except for the specified columns
for column in df.columns:
    if column not in exclude_columns and pd.api.types.is_numeric_dtype(df[column]):
        df[column] = df[column].round(0).astype(int)
```

### Feature Engineering

```python
# Price Range
df['Price_Range'] = df['Max_Price'] - df['Min_Price']
```

```python
# Average Price
df['Average_Price'] = (df['Min_Price'] + df['Max_Price']) / 2

# Price Variability Index
df['Price_Variability'] = (df['Frequent_price'] - df['Average_Price']) / df['Average_Price']

# Round Price_Variability to 4 decimal places
df['Price_Variability'] = df['Price_Variability'].round(4)
```

```python
# Weather Condition Indicator
df['Weather_Condition'] = df['weather'].apply(lambda x: 'Favorable' if 'clear' in x or 'sunny' in x else 'Unfavorable')
```
